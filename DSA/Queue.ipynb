### QUEUE

- FIFO (First In First Out)
- What are the items comes first They are removed First
- we may consider queue as a supply chain. In supplychain we try to push some items at one end of the pipe and we try to remove items from the other end of the pipe 
- dequeue - Front <- Rear - enqueue

*****Front :*****
- The front of the queue is where elements are removed from.
- It refers to the first element that was added to the queue (the earliest enqueued element).    

*****Rear :*****
- The rear of the queue is where elements are added to.
- It refers to the last element that was enqueued in the queue.   

Front → [10, 20, 30] ← Rear    
1) 10 is at the front (first to be dequeued).
2) 30 is at the rear (last element added, and new elements are added after it).

#### Operations 
- enqueue(x) : Add element x to the rear of the queue.
- dequeue(x) : Remove and return the front element of the queue.
- getFront() : Return the front element of the queue without removing it.
- getRear()  : Return the rear element of the queue without removing it.
- size()     : Return the number of elements in the queue.
- isEmpty()  : Check if the queue is empty.

### Applications of Queue

* Single Resource and Multiple Consumers (while taking tickets FCFS and new people added to line)
* Synchronization between slow and fast devices (like proceesor is fast and input of keybord is slow)
* In Operating Systems (Semaphores,,FCFS scheduling, spooling, buffers for devices like keyboard)
* In Computer Networks(Routers/Switches and Mail Queues)In Routers,which connect's multiple networks.having - sender network with high bandwidth, and with receiver network with lower bandwidth. sender send's the lot of data very quickly but receiver cannot consumes this data at that speed. Router have key internally, consumes the data and keep it up in the queue. whenever,low bandwidth network available it's take out the packet's here and going to send it to the receiver...this is how synchronization is maintained in. routers
* Variations : Deque, Priority queue, Doubly Ended Priority queue.

### Method3 Implemetation of queue with linkedlist

```python
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None


class MyQueue:
    def __init__(self):
        self.front = None
        self.rear = None
        self.sz = 0

    def size(self):
        return self.sz

    def isEmpty(self):
        return self.sz == 0

    def getFront(self):
        return self.front.key

    def getRear(self):
        return self.rear.key

    def enque(self, x):
        temp = Node(x)
        if self.rear == None:
            self.front = temp
        else:
            self.rear.next = temp

        self.rear = temp
        self.sz = self.sz + 1

    def deque(self):
        if self.front == None:
            return None
        else:
            res = self.front.key
            self.front = self.front.next
            if self.front == None:
                self.rear = None
            self.sz = self.sz - 1
            return res


q = MyQueue()
q.enque(10)
print(q.getFront(), q.getRear())
q.enque(20)
print(q.getFront(), q.getRear())
q.enque(30)
print(q.getFront(), q.getRear())
q.deque()
print(q.getFront(), q.getRear())
```

### Queue Implementation Using List

If we implement queue using list it going to take linear time for dequeue (pop(0)) operation .so, we don't want want this

```python
class MyQueue:

    def __init__(self, c):
        self.l = [None] * c
        self.cap = c
        self.size = 0
        self.front = 0

    def getFront(self):
        if self.size == 0:
            return None
        else:

            return self.l[self.front]

    def getRear(self):
        if self.size == 0:
            return None
        else:
            rear = (self.front + self.size - 1) % self.cap
            return self.l[rear]

    def enque(self, x):
        if self.size == self.cap:
            return
        else:
            rear = (self.front + self.size - 1) % self.cap
            rear = (rear + 1) % self.cap
            self.l[rear] = x

            self.size = self.size + 1

    def deque(self):
        if self.size == 0:
            return None
        else:
            res = self.l[self.front]
            self.front = (self.front + 1) % self.cap
            self.size = self.size - 1

            return res


# main
q = MyQueue(4)
q.enque(10)
print(q.getFront(), q.getRear())
q.enque(20)
print(q.getFront(), q.getRear())
q.enque(30)
print(q.getFront(), q.getRear())
q.enque(40)
print(q.getFront(), q.getRear())
q.deque()
print(q.getFront(), q.getRear())
q.deque()
print(q.getFront(), q.getRear())
q.enque(50)
print(q.getFront(), q.getRear())
```

The % self.cap ensures that once the index goes beyond the capacity (i.e., self.cap), it wraps back to 0 instead of going out of bounds.


### Implement Stack using Queue

There be taking two queues :  
- q1 : To keep the actual items
- q2 : To be used as an auxilary queue
    
*****Method1 :*****
1) Move Everything from q1 to q2
2) Enqueue x to q1
3) Move Everything back from q2 to q1

*****Method2 :*****
1) Enqueue x to q2
2) Move Everything from q1 to q2
3) Swap the variables q1 and q2 

```python
from collections import deque

class Stack :
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()
    def push(self,x):
        self.q2.append(x)
        
        while self.q1:
            self.q2.append(self.q1.popleft())
            
        self.q1,self.q2 = self.q2,self.q1
        
    def pop(self):
        if self.q1:
            self.q1.popleft()
    
    def top(self):
        if self.q1 :
            return self.q1[0]
        else:
            return None
    
    def size(self):
        return len(self.q1)
        
        
        
s = Stack()
s.push(3)
s.push(4)
s.push(5)
s.push(6)
print(s.size())
print(s.top())
s.pop()
print(s.top())
```
