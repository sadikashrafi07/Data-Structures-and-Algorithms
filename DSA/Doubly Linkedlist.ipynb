### Doubly Linked List

- In Doubly Linkedlist every node has reference to both Previous and Next nodes whereas, in singly linkedlist we have reference to Next node

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.prev = None
        self.next = None
        
head = Node(10)
temp1 = Node(20)
temp2 = Node(30)
head.next = temp1
temp1.prev = head
temp1.next = temp2
temp2.prev = temp1
```

### Advantages 
- Can be traversed in both directions(creating software for managing browser history th person visting multiple pages where he can go in both directions forward and backward..) 
- A given Node can be deleted in O(1) time with given reference/pointer to it.
- Insert/Delete before a given Node
- Insert/Delete from both ends in O(1) time by maintaing tail
### Disadvantages
- Extra Space for previous
- Code becomes more complex

### Insert at Begining of DLL

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.prev = None
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()
    
def insertBegin(head,x):
    temp = Node(x)
    if head != None:
        head.prev = temp
    temp.next = head
    return temp
        
head = None      
head = insertBegin(head,10)
head = insertBegin(head,20)
printlist(head)
```

### Insert at End of DLL

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.prev = None
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()

def insertEnd(head, x):
    temp = Node(x)
    if head == None:
        return temp
    curr = head
    while curr.next != None:
        curr = curr.next
    curr.next = temp
    temp.prev = curr
    return head
    
head = Node(10)
temp1 = Node(20)
temp2 = Node(30)
head.next = temp1
temp1.prev = head
temp1.next = temp2
temp2.prev = temp1

printlist(head)
insertEnd(head, 40)
printlist(head)

### Time Complexity theta(n)
```

### Delete head of DLL

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.prev = None
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()

def delHead(head):
    if head == None:
        return None
    elif head.next == None:
        return None
    else : 
        head = head.next
        head.prev = None
        return head
    
head = Node(10)
temp1 = Node(20)
temp2 = Node(30)
head.next = temp1
temp1.prev = head
temp1.next = temp2
temp2.prev = temp1

printlist(head)
head = delHead(head)
printlist(head)

### Time Complexity O(1)
### Auxilary Space O(1)
```

### Delete Last Node of DLL 

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.prev = None
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()

def delLast(head):
    if head == None:
        return None
    if head.next == None:
        return None
    curr = head
    while curr.next.next != None:
        curr = curr.next
    curr.next = None
    return head
    
head = Node(10)
temp1 = Node(20)
temp2 = Node(30)
head.next = temp1
temp1.prev = head
temp1.next = temp2
temp2.prev = temp1

printlist(head)
head = delLast(head)
printlist(head)

### Time Complexity O(n)
### Auxilary Space O(1)
```

### Reverse a Doubly linkedlist in Python

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.prev = None
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()

def reverseDLL(head):
    if head == None:
        return None
    if head.next == None:
        return head
    curr = head
    prev = None
    while curr != None:
        prev = curr
        curr.next,curr.prev = curr.prev,curr.next
        curr = curr.prev
    return prev 
    
head = Node(10)
temp1 = Node(20)
temp2 = Node(30)
head.next = temp1
temp1.prev = head
temp1.next = temp2
temp2.prev = temp1

printlist(head)
head = reverseDLL(head)
printlist(head)

### Time Complexity O(n)
### Auxilary Space O(1)
```
