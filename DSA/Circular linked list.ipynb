### Circular Linkedlist

- As,the name suggests it circular in nature
- The next of last nade is head
- In singly or doubly linkedlist the next of last node is None
- Circular linked list can be singly circular linkedlist or doubly circular linkedlist

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = head ### Considering the next of last node is head
```

##### Advantages
- we can traverse the whole list from any node
- Implementation of algorithm like round robin
- we can insert at begining and end just maintaining one tail reference/pointer (while inserting at the end we update the tail and while inserting at the begining we don't update the tail.we can do this in O(1).)
##### Disadvantages
- Implementation of operations becomes complex

- inserion at end in normal linkedlist is not possible in O(1) time with single pointer head. wheras in circular linkedlist is possible

### Circular linkedlist traversal

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printCircular(head):
    if head == None:
        return 
    print(head.key, end = " ")
    curr = head.next
    while curr != head:
        print(curr.key, end = " ")
        curr = curr.next
    
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = head ### Considering the next of last node is head
printCircular(head)
### Time Complexity in O(n) and space complexity is O(1)
```

- The duplicates don't affect the traversal, as you're comparing node references, not values.

### Insert at begining of Circular

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printCircular(head):
    if head == None:
        return 
    print(head.key, end = " ")
    curr = head.next
    while curr != head:
        print(curr.key, end = " ")
        curr = curr.next
    print() 
def insertBeg(head,x):
    temp = Node(x)
    if head == None:
        temp.next = temp
        return temp
    curr = head
    while curr.next != head:
        curr = curr.next
    curr.next = temp
    temp.next = head
    return temp
    
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = head ### Considering the next of last node is head
printCircular(head)
head = insertBeg(head,5)
printCircular(head)
### Time Complexity in O(n) and space complexity is O(1)
```

### Insert at End of Circular linkedlist

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printCircular(head):
    if head == None:
        return
    print(head.key, end = " ")
    curr = head.next
    while curr != head:
        print(curr.key, end = " ")
        curr = curr.next
    print() 
    
def insertEnd(head,x):
    temp = Node(x)
    if head == None:
        temp.next = temp
        return temp
    else :
        curr = head
        while curr.next != head:
            curr = curr.next
        curr.next = temp
        temp.next = head
        return head
    
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = head
printCircular(head)
insertEnd(head,40)
printCircular(head)

### Time Complexity O(n)
```

### Delete head of Circular Linkedlist

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printCircular(head):
    if head == None:
        return
    print(head.key, end = " ")
    curr = head.next
    while curr != head:
        print(curr.key, end = " ")
        curr = curr.next
    print() 

def delHead(head):
    if head == None:
        return None
    elif head.next == head :
        return None
    
    curr = head
    while curr.next != head :
        curr = curr.next
    curr.next = head.next
    return curr.next

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = head
printCircular(head)
head = delHead(head)
printCircular(head)
    
```

### Delete Kth Node of Circular LinkedList
