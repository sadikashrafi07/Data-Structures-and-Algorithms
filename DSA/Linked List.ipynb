### Traversing a linked list in a python

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.next = None


def printList(head):
    curr = head
    while curr!= None:
        print(curr.key, end=" ")
        curr = curr.next


head = Node(10)
head.next = Node(20)
head.next.next = Node(15)
head.next.next.next = Node(30)
printList(head)

### Time Complexity is theta(n) and space complexity is O(1)
```

### Search in Linked list

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def search(head,x):
    pos = 1
    curr = head
    while curr != None:
        if curr.key == x:
            return pos
        pos +=1
        curr = curr.next
    return -1
        
## Driver Code
head = Node(10)
head.next = Node(15)
head.next.next = Node(20)
head.next.next.next = Node(25)
x = 20
print(search(head,x))

## Worst Case O(n)
## Best Case O(1)
```

### Insert at the begining of linked list

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next
        
def insertBegin(head,key):
    temp = Node(key)
    temp.next = head
    return temp
    
### Driver Code
head = None
head = insertBegin(head,10)
head = insertBegin(head,20)
head = insertBegin(head,30)
printlist(head)
    
### Time Complexity O(1) and Space Complexity O(1)        
```

- When we create a new node `(temp = Node(key))` and point its next to the old head `(temp.next = head)`, the new node is the **new head** of the list.
- Since the function insertBegin creates a new head, we need to **return this new head** `(i.e., temp)` so that the calling code knows what the new head is.
- Without **return temp**, the head outside the function will **not be updated**, and we will lose track of the new head after inserting the node.
- temp return the **reference**. by assining it to head it becomes the head in driver code

### Insert at the end of the Linked List

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next

def insertEnd(head, key):
    if head == None:
        return Node(key)
    curr = head
    while curr.next != None:
        curr = curr.next
    curr.next = Node(key)
    return head

### Driver Code
head = None
head = insertEnd(head,10)
head = insertEnd(head,20)
head = insertEnd(head,30)
printlist(head)
                 
### Time Complexity O(n) and space Complexity O(1)
## If we initialize or maintain the last node we can insert at the end at O(1) time
```

1) How Traversal Works:
- You use a temporary pointer (curr) to traverse the list, starting from head.
- When we move "curr = curr.next", we are only moving the curr pointer, not the head. The head remains pointing to the first node.

- Here, `head` is **None**.  
- Inside the function:
  - You create a **new node** and `return` it.  
- The caller (`head = ...`) assigns that new node as the new head.  
-  **If you didn’t return it**, your external `head` would still be `None` — the list would be lost.

### Insert at given position in singly Linked List

- To insert the data at given position we should required the previous position of an element
- If we inserting an element at pos 4 we required pos 3 to insert. to get pos 3 we iterate 2 times from pos 1. then general expression becomes (pos-2)

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.next = None

def printlist(head):
    curr = head
    while curr is not None:
        print(curr.key, end=" ")
        curr = curr.next

def insertPos(head, key, pos):
    temp = Node(key)
    if pos == 1:             ## While insert at begining
        temp.next = head     ##head is reference of previous node (## Link the new node's next pointer to the current head)
        return temp          ## Return the new node as the new head
    curr = head              ## While iterating store head in curr (## Initialize a pointer (curr) to traverse the list)
    for i in range(pos-2):
        curr = curr.next
        if curr == None:
            return head
    temp.next = curr.next    ### New Node points to the Next Node
    curr.next = temp         ###Current node points to the new node
    return head


### Driver Code
head = Node(10)
head.next = Node(20)
head.next.next = Node(15)
head.next.next.next = Node(30)


# Insert a new node with key 25 at position 3
head = insertPos(head, 25, 3)

# Print the list
printlist(head)  
```

- curr: Points to the current node in the list (the node before the insertion point).
- curr.next: Points to the node that comes after curr in the list.
- temp: The new node that is being inserted into the list.  
****Note :**** We first link the new node (temp) to the next node in the list to ensure it maintains a connection to what follows. Then, we link the previous node (curr) to the new node, effectively inserting it into the list without losing any existing connections.

### Delete Last Node of Linked List

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
        
def delLast(head):
    if head == None:
        return None
    if head.next == None:
        return None
    curr = head
    while curr.next.next != None:   ## We are iterating upto before end
        curr = curr.next
    curr.next = None
    return head
    
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head = delLast(head)
printlist(head)

### Time Complexity theta(n) and Space Complexity is theta(1)
```

### Sorted insert in a linked list

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
        
def sortedInsert(head, x):
    temp = Node(x)
    if head == None:
        return temp
    elif x < head.key:
        temp.next = head
        return temp
    else :
        curr = head
        while curr.next != None and curr.next.key < x:
            curr = curr.next
        temp.next = curr.next
        curr.next = temp
        return head

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head = sortedInsert(head, 35)
printlist(head)

### Time Complexity theta(n) and Space Complexity is theta(1)
```

### Middle of Linked List

```python
### Naive approach 

class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
        
def printMiddle(head):
    if head == None:
        return
    count = 0
    curr = head
    while curr:
        curr = curr.next
        count += 1
    curr = head
    for i in range(count//2):
        curr = curr.next
    return curr.key
        
head = Node(10)
head.next = Node(5)
head.next.next = Node(20)
head.next.next.next = Node(15)
head.next.next.next.next = Node(25)
printlist(head)
printMiddle(head)
```

```python
### Efficient approach

class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
        
def printMiddle(head):
    if head == None:
        return "List is empty"
    slow = head
    fast = head
    while fast != None and fast.next != None:
        slow = slow.next
        fast = fast.next.next
    return slow.key
    

        
head = Node(10)
head.next = Node(5)
head.next.next = Node(20)
head.next.next.next = Node(15)
head.next.next.next.next = Node(25)
printlist(head)
printMiddle(head)
```

1)Why Check fast and fast.next?     
     
*****fast***** is not None:
- This condition ensures that the fast pointer has not reached the end of the list. Once fast is None, the list traversal is complete. 

*****fast.next***** is not None:  
- Since fast moves two steps forward at a time (fast = fast.next.next), you also need to check that fast.next exists.
- If fast.next is None, fast.next.next would raise an error, as you'd be trying to access None.next, which is invalid.

### Remove Duplicates from a Sorted List

```python

class Node:
    def __init__(self,key):
        self.key = key
        self.next = None

def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()

def removeDups(head):
    curr = head
    while curr != None and curr.next != None:
        if curr.key == curr.next.key:
            curr.next = curr.next.next
        else:
            curr = curr.next       

        
head = Node(10)
head.next = Node(20)
head.next.next = Node(20)
head.next.next.next = Node(30)
head.next.next.next.next = Node(30)
printlist(head)
removeDups(head)
printlist(head)

## Time Complexity O(n) no:of nodes in input linkedlist and auxilay space O(1)
```

### Reverse a Linkedlist in Python

```python

class Node:
    def __init__(self,key):
        self.key = key
        self.next = None

def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()

def reverselists(head):
    stack = []
    curr = head
    while curr is not None:
        stack.append(curr.key)
        curr = curr.next
    curr = head
    while curr is not None:
        curr.key = stack.pop()
        curr = curr.next
    return head
      

        
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
printlist(head)
reverselists(head)
# Print the list after reversing
printlist(head)
## Time Complexity O(n) and auxilay space required list
```

```python
### Reversing the links
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None

def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()

def reverselists(head):
    curr = head
    prev = None   ## We should make first element to direct towards None
    while curr != None:
        next = curr.next   ## We should hold first next element to don't miss
        curr.next = prev   ## We are redirecting link to next insted to previous which is none. Now we have hold of second element also
        prev = curr        ## Here current becomes previous
        curr = next        ## curr and next both points the same 1) which we stored (next = curr.next)
    return prev            ## at end curr and next becomes none and we take prev as new head of linkedlist
      

        
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
printlist(head)

# Reverse the linked list
head = reverselists(head)

# Print the list after reversing
printlist(head)

# Time Complexity: O(n) 
# Auxiliary Space: O(1)
```

By doing head = reverselists(head), you are updating the head to point to the new starting node of the reversed list.     

*****If you don't update head:******
- If you do not store the result in head, the original head pointer will still point to the first node of the old (non-reversed) list, and you won't be able to access the reversed list.

### Recursive reverse a linked-list Part 1

```python
class Node:
    def __init__(self,key):
        self.key = key
        self.next = None
        
def printlist(head):
    curr = head
    while curr != None:
        print(curr.key, end = " ")
        curr = curr.next
    print()
    
def reverselist(head):
    if head == None:          ## If we reach head to none
        return head
    if head.next == None:     ## If there is only one element 
        return head 
    rest_head = reverselist(head.next) ## We are making recursive call from second element and created linkedlist from reverse except head and got new head as resthead
    rest_tail = head.next   ##To link the head we are finding next element of head and considered as rest_tail
    rest_tail.next = head   ## Now that head's Next element was linked to head (head <- next)
    head.next = None        ## The end Should be None so, we linking None Next to head (None <- head)
    return rest_head
        
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
printlist(head)
head = reverselist(head)
printlist(head)
```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.next = None
        
def printlist(head):
    curr = head
    while curr is not None:
        print(curr.key, end=" ")
        curr = curr.next
    print()

def reverselist(curr, prev=None):
    if curr is None:  # Base case: If we reach the end of the list, return the new head
        return prev
    next = curr.next  # Store the next node
    curr.next = prev  # Reverse the current node's pointer
    return reverselist(next, curr)  # Recursive call with next node and current node as the new previous node


head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

# Print original list
print("Original list:")
printlist(head)

# Reverse the list
head = reverselist(head)

# Print reversed list
print("Reversed list:")
printlist(head)
```
