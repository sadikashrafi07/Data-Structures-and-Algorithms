```python
def find_num(l,x):
    for i in range(len(l)):
        if l[i] == x:
            return i
    return -1
        
l = [10,20,30,40,50,60]
x = 20
print(find_num(l,x))

### Time complexity of this code is o(n) it is linear search it is searching with each element
```

### Binary Search

```python
def bsearch(l,x):
    low = 0
    high = len(l)-1
    while low <= high:
        mid = (low + high) // 2
        if l[mid] == x:    # If x is equal to the middle element, return the index
            return mid
        elif l[mid] < x :  # If x is greater than the middle element, adjust the lower bound
            low = mid + 1
        else :             # If x is less than the middle element, adjust the upper bound
            high = mid -1
    return -1
            
l = [10,20,30,40,50,60]
x = 20
print(bsearch(l,x))        
### Time complexity of this code is o(log(n))
### Aux space is o(1)
```

```python
def bsearch(l,x,low,high):
    if low>high:
        return -1
    mid = (low + high) // 2
    
    if l[mid] == x:
        return mid
    elif l[mid] < x:
        return bsearch(l,x,mid+1,high)
    else :
        return bsearch(l,x,low,mid-1)
    
l = [10, 20, 30, 40, 50, 60]
x = 20
low = 0
high = len(l) - 1
print(bsearch(l,x,low,high))  

### Time complexity of this code is o(log(n))
### Aux space is o(log(n))
### Not Preffered
```

### Index of First Occurence

```python
def firstOccurence(l, x):
    low=0 
    high=len(l)-1
    while(low<=high):
        mid=(low+high)//2
        if(x>l[mid]):
            low=mid+1 
        elif(x<l[mid]):
            high=mid-1 
        else:
            if(mid==0 or l[mid-1]!=l[mid]):
                return mid 
            else:
                high=mid-1
    
    return -1

l= [1,10,10,10,20,20,40]
x = 20
print(firstOccurence(l,x))   
```

If neither of the above conditions is true (i.e., mid is not 0 and the previous element is the same as l[mid]), it means that x has appeared earlier in the list. Therefore, the algorithm needs to continue searching in the left half of the list, so it adjusts high = mid - 1 to narrow down the search.

### Index of last Occurence

```python
def lastOccur(l,x):

    for i in reversed(range(len(l))):

        if l[i] == x:
            return i
    return -1

l = [10,15,20,20,40,40]

print(20,lastOccur(l, 20))
print(40,lastOccur(l, 40))

### Time Complexity o(n)  linear search
```

### Count Occurences in sorted array

```python
def firstOccurence(l, x):
    low=0 
    high=len(l)-1
    while(low<=high):
        mid=(low+high)//2
        if(x>l[mid]):
            low=mid+1 
        elif(x<l[mid]):
            high=mid-1 
        else:
            if(mid==0 or l[mid-1]!=l[mid]):
                return mid 
            else:
                high=mid-1
    
    return -1

def lastOccurence(l, x):
    low=0 
    high=len(l)-1
    while(low<=high):
        mid=(low+high)//2
        if(x>l[mid]):
            low=mid+1 
        elif(x<l[mid]):
            high=mid-1 
        else:
            if(mid==len(l)-1 or l[mid]!=l[mid+1]):
                return mid 
            else:
                low = mid +1
    
    return -1

def countOcc(l,x):
    first = firstOccurence(l,x)
    if first == -1:
        return 0
    else :
        return lastOccurence(l,x) - first + 1
    
l = [10,20,20,20,20,20,30,30]
x = 20
print(countOcc(l,x))

### Time Complexity : O(log(n))+O(log(n)) = O(log(n))
```

### Count 1's in a Sorted Binary List

```python
def countOnes(l, n):
	low = 0
	high = n - 1
	while (low <= high): # get the middle index
		mid = (low + high) // 2

		if (l[mid] < 1):
			high = mid - 1

		elif(l[mid] > 1):
			low = mid + 1
		else:

			if (mid == n - 1 or l[mid + 1] != 1):
				return mid + 1
			else:
				low = mid + 1

	return 0


if __name__ == '__main__':

	arr = [1, 1, 1, 1, 0, 0, 0]
	n = len(arr)

	print("Count of 1's in given array is ", countOnes(arr, n))
```

### Square root

```python
def sqrootfloor(x):
    i = 1
    while i*i <= x:
        i += 1
    return i-1

x = 25
print(sqrootfloor(x))

### Time complexity O(sqrt(n))
```
